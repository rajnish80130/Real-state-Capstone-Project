```python
import re
df['society']= df['society'].apply(lambda name :re.sub(r'\d+(\.\d+)?\s?★','',str(name)).strip()).str.lower()
```

```python
df['society'].value_counts().shape
```

```python
df['society']=df['society'].str.replace('nan','Independent')
```

```python
#price
df['price'].value_counts()
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
df.shape
```

```python
df.head()
```

```python
df['price'].isnull().sum()
```

```python
def treat_price(x):
  if type(x)==float:
    return x
  else:
    if x[1] == 'Lac':
      return round(float(x[0])/100,2)
    else:
      return round(float(x[0]),2)
```

```python
df['price']=df['price'].str.split(' ').apply(treat_price)
```

```python
df.sample(5)
```

```python
df.dropna(subset=['price'],inplace=True)
```

```python
df.shape
```

```python
df['price'].isnull().sum()
```

```python
#price_per_sqft
df['price_per_sqft'].value_counts()
```

```python
df['price_per_sqft'].isnull().sum()
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype(int)
```

```python
df.head()
```

```python
df['area'].isnull().sum()
```

```python
df['area']=df['area'].str.split('sq').str.get(0).str.split('(').str.get(1).str.strip().astype(int)
```
